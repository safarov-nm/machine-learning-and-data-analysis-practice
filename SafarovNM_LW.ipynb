<p style="font-family: Times New Roman;">В этом пункте начинается подготовка данных для кластеризации. Сначала мы извлекаем целевые метки из набора данных, которые находятся в столбце outcome, и сохраняем их в переменную y. Эти метки представляют собой истинные значения, с которыми мы будем сравнивать результаты кластеризации. Признаки для кластеризации, которые содержат информацию о каждом объекте (строке данных), сохраняются в переменной X. Из этой матрицы признаков мы удаляем столбец с целевой переменной (outcome). Затем данные масштабируются с помощью стандартизации, где каждый признак приводится к среднему значению 0 и стандартному отклонению 1. Это нужно, чтобы избежать искажений из-за различий в масштабах признаков, поскольку алгоритмы кластеризации чувствительны к этим различиям. Для стандартизации используется класс StandardScaler из библиотеки sklearn, который вычисляет среднее значение и стандартное отклонение для каждого признака и затем применяет к данным преобразование.</p>

```python
y = data['outcome']  # Целевая переменная (истинные метки)
X = data.drop(columns=['outcome'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

<p style="font-family: Times New Roman;"><b>Пункт №1.</b> Применение PCA.</p>

<p style="font-family: Times New Roman;">После подготовки данных их нужно проецировать в двумерное пространство для визуализации, чтобы можно было лучше понять, как алгоритмы кластеризации распределяют объекты. Для этого используется метод главных компонент (PCA), который снижает размерность данных. В данном случае мы выбираем две компоненты для отображения данных в 2D. Это делается с помощью класса PCA из библиотеки sklearn.decomposition. Метод fit_transform применяет преобразование к данным, и мы получаем новые данные с двумя признаками, которые можно визуализировать. Процесс PCA позволяет уменьшить количество признаков, сохраняя при этом как можно больше информации о различиях между объектами.</p>

<p style="font-family: Times New Roman;"><b>Пункт №2.</b> Кластеризация.</p>

<p style="font-family: Times New Roman;">Здесь мы применяем три разных метода кластеризации, чтобы проверить, какой из них лучше разделяет объекты на группы. Первый метод — это KMeans, который выполняет кластеризацию, пытаясь разделить данные на фиксированное количество кластеров (в данном случае на два). Мы создаем объект KMeans, указываем количество кластеров как 2 и запускаем метод fit_predict, который обучает модель и предсказывает метки кластеров для каждого объекта в данных. Второй метод — это DBSCAN (Density-Based Spatial Clustering of Applications with Noise), который использует плотность данных для разделения объектов на кластеры. DBSCAN имеет два важных параметра: eps (максимальное расстояние между соседними точками в одном кластере) и min_samples (минимальное количество точек, чтобы считать область плотной). Этот алгоритм также может помечать точки как выбросы (метка -1). Третий метод — это иерархическая кластеризация (Agglomerative Clustering), которая строит дерево кластеров, начиная с каждого объекта как отдельного кластера и постепенно объединяя их в более крупные кластеры. Мы выбираем два кластера для иерархической кластеризации и получаем метки кластеров для каждого объекта. Все три алгоритма создают метки кластеров, которые будут использованы для оценки качества кластеризации.</p>

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled) 

agglo = AgglomerativeClustering(n_clusters=2) 
agglo_labels = agglo.fit_predict(X_scaled) 
```

<p style="font-family: Times New Roman;"><b>Пункт №3.</b> Оценка качества кластеризации (ARI).</p>

<p style="font-family: Times New Roman;">В этом пункте мы оцениваем качество кластеризации с помощью Adjusted Rand Index (ARI). ARI измеряет схожесть между истинными метками классов и метками кластеров, созданными алгоритмами кластеризации. Это значение варьируется от -1 до 1, где 1 означает полное совпадение, а значения близкие к 0 или отрицательные указывают на то, что кластеризация не имеет никакой связи с истинными метками. Для каждой из трех моделей мы вычисляем ARI с помощью функции adjusted_rand_score из библиотеки sklearn.metrics и выводим результаты. ARI для KMeans равен 0.1497, для DBSCAN — 0.1342, и для Agglomerative Clustering — 0.0491. Все эти значения достаточно низкие, что указывает на слабую корреляцию между кластеризацией и истинными метками, но наименьшее значение наблюдается для иерархической кластеризации, что свидетельствует о плохом разделении объектов.</p>

```python
ari_kmeans = adjusted_rand_score(y, kmeans_labels)  # Оценка ARI для KMeans
ari_dbscan = adjusted_rand_score(y, dbscan_labels)  # Оценка ARI для DBSCAN
ari_agglo = adjusted_rand_score(y, agglo_labels)  # Оценка ARI для Agglomerative Clustering

# Выводим результаты ARI
print(f"ARI for KMeans: {ari_kmeans:.4f}")
print(f"ARI for DBSCAN: {ari_dbscan:.4f}")
print(f"ARI for Agglomerative Clustering: {ari_agglo:.4f}")
```

<p style="font-family: Times New Roman;"><b>Пункт №4.</b> Оценка качества кластеризации (Silhouette Score).</p>

<p style="font-family: Times New Roman;">Здесь мы оцениваем качество кластеризации с помощью Silhouette Score. Этот показатель измеряет, насколько хорошо каждый объект подходит к своему кластеру, сравнивая его с ближайшим соседним кластером. Значение этого показателя варьируется от -1 до 1, где 1 означает, что объект хорошо кластеризован, 0 — что объект находится на границе двух кластеров, а -1 — что объект был ошибочно отнесен к своему кластеру. Для каждого из алгоритмов мы рассчитываем Silhouette Score, используя функцию silhouette_score из библиотеки sklearn.metrics. Для DBSCAN исключаются выбросы (метки -1), так как они не принадлежат никакому кластеру. Для KMeans Silhouette Score равен 0.1516, для DBSCAN — 0.5058, а для Agglomerative Clustering — 0.1298. Наибольшее значение Silhouette Score наблюдается для DBSCAN, что указывает на лучшую сепарацию кластеров по сравнению с другими алгоритмами.</p>

```python
# Для DBSCAN исключаем выбросы (-1) и соответствующие точки
dbscan_labels_filtered = dbscan_labels[dbscan_labels != -1]  # Исключаем выбросы
X_scaled_filtered = X_scaled[dbscan_labels != -1]  # Исключаем данные с меткой -1

# Рассчитываем Silhouette Score для каждой модели
silhouette_kmeans = silhouette_score(X_scaled, kmeans_labels)  # Для KMeans
silhouette_dbscan = silhouette_score(X_scaled_filtered, dbscan_labels_filtered)  # Для DBSCAN (без выбросов)
silhouette_agglo = silhouette_score(X_scaled, agglo_labels)  # Для Agglomerative Clustering

# Выводим результаты Silhouette Score
print(f"Silhouette Score for KMeans: {silhouette_kmeans:.4f}")
print(f"Silhouette Score for DBSCAN: {silhouette_dbscan:.4f}")
print(f"Silhouette Score for Agglomerative Clustering: {silhouette_agglo:.4f}")
```
